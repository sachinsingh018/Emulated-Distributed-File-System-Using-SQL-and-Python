```python
def enhancedreducer(table,attributelisttodisplay,attributestosearchby,attributevaluesforsearch):
    list_all=[]
    for i in range(len(attributestosearchby)):
        df=enhancedmapper(table,i,attributestosearchby[i],attributevaluesforsearch[i])
        list_all.append(df)
    df1=list_all[0]
    df2=list_all[1]
    D=pd.merge(df1, df2, how ='inner', on =['DR_NO', 'DR_NO'])
    for j in range(2,len(list_all)):
        df=list_all[j]
        D=pd.erge(D,df,how='inner',on=['DR_NO', 'DR_NO'])
    D=D[attributelisttodisplay]
    print(D)
        
```

```python
def cat(table,partitions):
    sentence="select * from "+table+str(0)
    df1=pd.read_sql(sentence,con=connection)
    
    sentences="select * from "+table+str(1)
    df2=pd.read_sql(sentences,con=connection)

    dfinitial=pd.concat([df1,df2],axis=1)
    for i in range(2,partitions):
        sentence="select * from table"+str(i)
        dfe=pd.read_sql(sentence,con=connection)
        dfinitial=pd.concat([dfinitial,dfe],axis=1)
    print(dfinitial)
```
